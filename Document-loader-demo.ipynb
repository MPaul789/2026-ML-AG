Import the PDF Loader

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("sample.pdf")
documents = loader.load()
```

Check the Type of Output

```python
print(type(documents))
```

Check Number of Pages/Documents

```python
print(len(documents))
```

Access First Document Object

```python
print(documents[0])
```

Access Only Page Content

```python
print(documents[0].page_content)
```

Access Metadata

```python
print(documents[0].metadata)
```

Understand Document Structure

```python
print(type(documents[0]))
```

Using Load

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("sample.pdf")

documents = loader.load()

print(type(documents))
print(len(documents))
```

Using Lazy_load

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("sample.pdf")

documents = loader.lazy_load()

print(type(documents))
```

A generator is a special type of Python object that produces values one at a time only when needed, instead of storing all values in memory at once.
