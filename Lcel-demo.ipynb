Importing LCEL components

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline

from langchain_core.runnables import (
    RunnableLambda,
    RunnablePassthrough,
    RunnableParallel
)
```

Create Your First Prompt Template

```python
prompt = ChatPromptTemplate.from_template(
    "Explain {topic} in simple words."
)
```

Testing the Prompt

```python
formatted_prompt = prompt.invoke(
    {
        "topic": "LCEL"
    }
)

print(formatted_prompt)
```

```python
pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    max_new_tokens=100
)
```

```python
llm = HuggingFacePipeline(
    pipeline=pipe
)
```

Create Your First LCEL Chain

```python
chain = prompt | llm
```

Invoke the LCEL Chain

```python
response = chain.invoke(
    {
        "topic": "LCEL"
    }
)

print(response)
```

Create Chain with Output Parser

```python
parser_chain = prompt | llm | StrOutputParser()
```

Invoke the Parser Chain

```python
response = parser_chain.invoke(
    {
        "topic": "LCEL"
    }
)

print(response)
```

Create a RunnableLambda

```python
def convert_to_upper(text):

    return text.upper()
```

```python
uppercase_runnable = RunnableLambda(
    convert_to_upper
)
```

```python
result = uppercase_runnable.invoke(
    "hello langchain"
)

print(result)
```

Create LCEL Chain with RunnableLambda

```python
lambda_chain = (
    prompt
    | llm
    | StrOutputParser()
    | uppercase_runnable
)
```

Invoke the RunnableLambda Chain

```python
response = lambda_chain.invoke(
    {
        "topic": "LCEL"
    }
)

print(response)
```

Create RunnablePassthrough Chain

```python
passthrough_chain = (
    RunnablePassthrough()
)
```

Test RunnablePassthrough

```python
result = passthrough_chain.invoke(
    "Hello LCEL"
)

print(result)
```

Create RunnableParallel

```python
parallel_chain = RunnableParallel(
    original = RunnablePassthrough(),

    uppercase = uppercase_runnable
)
```

Test RunnableParallel

```python
result = parallel_chain.invoke(
    "hello langchain"
)

print(result)
```
